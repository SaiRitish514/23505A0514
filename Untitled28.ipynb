```python
import tensorflow as tf

train_dir = "/content/drive/MyDrive/Combined Dataset/train"

IMG_SIZE = (224,224)
BATCH_SIZE = 32

train_data = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_data = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_data.class_names
print("Classes:", class_names)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_data = train_data.prefetch(buffer_size=AUTOTUNE)
val_data = val_data.prefetch(buffer_size=AUTOTUNE)
```

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
])
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))

for layer in base_model.layers[:-20]:
    layer.trainable = False

x = data_augmentation(base_model.output)
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(4, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=Adam(1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
labels = np.concatenate([y for x, y in train_data], axis=0)
weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weights = dict(enumerate(weights))
print(class_weights)
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    class_weight=class_weights
)
```

```python
for layer in base_model.layers[-30:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history2 = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'] + history2.history['accuracy'])
plt.plot(history.history['val_accuracy'] + history2.history['val_accuracy'])
plt.legend(['Train','Val'])
plt.title("Accuracy")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

y_true = []
y_pred = []

for x, y in val_data:
    pred = model.predict(x)
    y_pred.extend(np.argmax(pred, axis=1))
    y_true.extend(y.numpy())

cm = confusion_matrix(y_true, y_pred)

sns.heatmap(cm, annot=True, fmt='d')
plt.show()

print(classification_report(y_true, y_pred, target_names=class_names))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os

# 🔹 Dataset folder
folder = "/content/drive/MyDrive/Combined Dataset/test"

# 🔹 Automatically pick 1 image from any class
for class_name in os.listdir(folder):
    class_path = os.path.join(folder, class_name)

    if not os.path.isdir(class_path):
        continue

    files = os.listdir(class_path)
    if len(files) == 0:
        continue

    img_path = os.path.join(class_path, files[0])
    break

print("✅ Using image:", img_path)

# 🔹 Grad-CAM function
def gradcam(img_path):
    img = cv2.imread(img_path)

    if img is None:
        print("❌ Image not found")
        return

    img = cv2.resize(img, (224,224))

    arr = tf.keras.applications.resnet50.preprocess_input(img)
    arr = np.expand_dims(arr, 0)

    # 🔥 Correct conv layer for ResNet50
    grad_model = tf.keras.models.Model(
        [model.inputs],
        [model.get_layer("conv5_block3_out").output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(arr)
        idx = tf.argmax(preds[0])
        loss = preds[:, idx]

    grads = tape.gradient(loss, conv_out)
    pooled = tf.reduce_mean(grads, axis=(0,1,2))

    conv_out = conv_out[0]
    heatmap = conv_out @ pooled[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    heatmap = np.maximum(heatmap, 0)
    heatmap = heatmap / (np.max(heatmap) + 1e-8)
    heatmap = cv2.resize(heatmap, (224,224))

    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    overlay = cv2.addWeighted(img, 0.6, heatmap, 0.4, 0)

    plt.figure(figsize=(6,6))
    plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
    plt.title("Grad-CAM")
    plt.axis('off')
    plt.show()

# 🔹 Prediction + Stage
def predict_stage(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224,224))

    arr = tf.keras.applications.resnet50.preprocess_input(img)
    arr = np.expand_dims(arr, 0)

    pred = model.predict(arr)
    idx = np.argmax(pred)
    conf = np.max(pred)

    print("\n🧠 Prediction:", class_names[idx])
    print("📊 Confidence:", round(conf*100,2), "%")

# 🔹 RUN BOTH
predict_stage(img_path)
gradcam(img_path)
```

```python
import tensorflow as tf

img_size = (224, 224)
batch_size = 32

train_data = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=img_size,
    batch_size=batch_size
)

test_data = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    image_size=img_size,
    batch_size=batch_size
)

print("Classes:", train_data.class_names)
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(224,224,3)),

    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),

    # 👇 4 classes kabatti 4 neurons
    layers.Dense(4, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(train_data, epochs=10, validation_data=test_data)
```
